## StandOut Section

```python
df.corr()
```

```python
df.plot(x= 'Quantity', y= 'gross income', kind= 'scatter')
```

```python
df.plot(x= 'Quantity', y= 'gross income', kind= 'line')
```

```python
df.plot(x= 'Unit price', y= 'Quantity', kind= 'scatter')
```

```python
df.plot(x= 'cogs', y= 'gross income', kind= 'line')
```

```python
df.plot(x= 'cogs', y= 'Tax 5%', kind= 'line')
```

#### Using the corr() method, the table above shows lots of numbers that represent how well the relationship is between two columns(variables).
- There is a fair relationship between the quantity of goods sold and the gross income with a correlation coefficient of 0.705510, which means that as the number of products purchased by customers increases, the gross income will probably increase as well, as shown in the line and scatter plot above.

- There is no relationship between the price of product(Unit price) and the number of products purchased by the customer with a correlation coefficient of 0.010778. In other words, the price of product is not affected by the quantity of goods purchased by the customer as seen in the scatter plot above.

- There is a perfect linear relationship between the cost of goods sold and the gross income with a correlation coefficient of 1. This means that more income will always be generated as the  cost of goods sold is increased. as shown in the line plot above.

- There is a perfect relationship between the tax fee for customer buying(Tax 5%) and the cost of goods sold with a correlation coefficient of 1. This simply means that the tax increases as the cost of goods sold is increased as displayed in the line plot above.
